### 3. Key ideas + why it’s interesting <a class="anchor" id="bullet3"></a>

A big concept here is **defensive spacing**: measuring how “spread out” the defense is.

The repo’s README discusses using a convex hull approach (from `scipy.spatial.ConvexHull`) to quantify spacing.

In plain English: if defenders are pulled farther apart, it can open lanes for drives and passes — *spacing* becomes a measurable thing.

### 4. Example Plot From the Repo <a class="anchor" id="bullet4"></a>

 Below, I will load **an image plot that is included in the repository** (from the `examples/` folder on GitHub) and display it inside this notebook.

```python
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

# A plot image stored in the repo's examples/ folder (used in the README)
img_url = "https://raw.githubusercontent.com/christopherjenness/NBA-player-movement/master/examples/DefensiveSpacing.png"

resp = requests.get(img_url)
resp.raise_for_status()

img = Image.open(BytesIO(resp.content))

plt.figure(figsize=(10, 6))
plt.imshow(img)
plt.axis("off")
plt.title("Plot from NBA-player-movement repo: DefensiveSpacing.png")
plt.show()
```
