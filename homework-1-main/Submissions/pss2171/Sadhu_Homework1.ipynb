```python
industry = ["Supply Chain", "E-commerce", "Cybersecurity", 
            "Gaming", "Hospitality", "Human Resources", "Legal",
            "Entertainment", "Energy", "Healthcare", "Finance",
           "Education", "Customer Service", "Retail", "Transportation",
           "Manufacturing", "Real Estate", "Agriculture"]
usecase_freq = [1, 1, 2, 1, 1, 1, 1, 1, 1, 4, 1, 1, 1, 1, 1, 1, 1, 1]

plt.figure(figsize=(12, 5))
plt.bar(industry, usecase_freq)
plt.ylabel("Number of Documented Use Cases")
plt.title("AI Agent Use Cases Per Industry")

plt.xticks(rotation=90, ha="right")
plt.tight_layout()
plt.show
```
