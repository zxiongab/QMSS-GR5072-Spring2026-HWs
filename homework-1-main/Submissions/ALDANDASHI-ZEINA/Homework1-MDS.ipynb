### 1. Project Overview <a class="anchor" id="section1"></a>

This GitHub project is a demo application built with Streamlit that visualizes Uber pickup data in New York City. It allows users to explore when and where Uber pickups occur by interacting with maps and time-based filters. The project is designed to showcase how Python can be used to turn real datasets into interactive and user-friendly data apps.

### 2. Why Its Interesting <a class="anchor" id="section2"></a>
This project is interesting because it turns a **large** and **complex** real-world dataset into an interactive visualization that is easy to explore and understand. Instead of relying on static charts, users can actively filter the data to uncover patterns in Uber pickup behavior across time and space. This demonstrates how data visualization tools can make insights more accessible to both technical and non-technical audiences.

###       2.1. Key Figures <a class="anchor" id="section2-1"></a>
- An interactive geographic map showing the density and distribution of Uber pickups in NYC  
- Dynamic time filters that help users identify peak pickup hours and daily trends  
- A simple Streamlit interface that demonstrates how Python can power interactive data applications  


### 3. Plot / Figure <a class="anchor" id="section2"></a>
  **(*Figure 1*:)** This visualization shows how Uber pickups are distributed across New York City during a specific hour of the day. The map highlights areas with higher pickup density, while the bar chart below breaks down ride frequency by minute. This combination allows users to examine both where and when Uber demand is highest.

### 4. Github & Github Reflection <a class="anchor" id="section4"></a>
When viewing the commit history for the Python script file, the code looked familiar because it was stored as plain text and displayed exactly as it was written. In contrast, the Jupyter Notebook commit history appeared less familiar because the `.ipynb` file is stored as a JSON object that includes metadata, cell structure, and output information in addition to the code itself. This makes the notebook history harder to read compared to a standard Python script, even though the underlying code is still present.
