# Plot <a id="plot"></a>


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)

plt.figure()
plt.plot(x, y)
plt.title("Example plot made with matplotlib")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```
