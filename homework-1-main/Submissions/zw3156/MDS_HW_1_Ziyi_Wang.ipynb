###1.2. About the Dataset <a class="anchor" id="bullet1-2"></a>

The Iris dataset is a classic dataset commonly used in machine learning and statistics. It contains measurements of iris flowers, including sepal length, sepal width, petal length, and petal width.

Each observation belongs to one of three species, ***Iris setosa***, ___Iris versicolor___, or ***Iris virginica***. Because the dataset is small, clean, and well-structured, it is often used for teaching and demonstrating classification algorithms.

##2. Visualization Example <a class="anchor" id="bullet2"></a>

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target

plt.figure()
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'])
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Sepal Width (Iris Dataset)')
plt.show()
```

The scatter plot above visualizes the relationship between sepal length and sepal width in the Iris dataset. Each point represents one flower observation, and the plot helps illustrate how these two features vary across the dataset.
