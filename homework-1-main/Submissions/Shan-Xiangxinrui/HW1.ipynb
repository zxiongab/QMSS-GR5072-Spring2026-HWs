## 4. Plot <a class="anchor" id="bullet4"></a>

The following is a plot.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("Sine Wave Example")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```
