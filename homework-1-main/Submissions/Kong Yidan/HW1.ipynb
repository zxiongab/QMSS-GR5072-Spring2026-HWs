## 1. Project Overview <a class="anchor" id="bullet1"></a>

**Project name:** deck.gl  
**Link to the GitHub repo:** https://github.com/visgl/deck.gl

deck.gl is an open source library for data visualization on the web. It is designed for large datasets and uses GPU accelerated rendering so interactive visuals stay fast. The main idea is to map data into visual layers like points, lines, text, and polygons, then combine layers to build a complete view. It also supports interaction like picking and highlighting, which makes it useful for maps and dashboards.

## 2. What I Found Interesting <a class="anchor" id="bullet2"></a>

What I like about deck.gl is how practical it feels. It focuses on making large datasets smooth to explore in the browser, which is harder than it looks. I also like the layer concept because it keeps the workflow clear and modular. To me, this is *a clean example of turning structured data into something people can actually interact with*. The biggest takeaway is **performance and interactivity are treated as first class goals**.
