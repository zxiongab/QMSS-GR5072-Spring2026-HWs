# Program overview
## Background

The **ESPN API GitHub project** (the `espn_api` package) provides tools to access data from ESPN's Fantasy APIs in a more convenient and *programmatic* way.  
Instead of manually checking scores or standings on the ESPN website, users can write Python code to:

- download league, team, and player information
- analyze fantasy performance over time
- build custom dashboards or alerts for their fantasy leagues

Here is the link to the project on GitHub:  
**[ESPN API GitHub repository](https://github.com/cwendt94/espn-api)**

In this notebook, I will:
- briefly describe what the project does
- show an example plot that could be created using data from this kind of API
- summarize a few **important points** about why this project is useful.

The code will be down below with the sample problem.

```python
import matplotlib.pyplot as plt

# Hypothetical weekly fantasy scores for one team
weeks = [1, 2, 3, 4, 5, 6]
scores = [95, 103, 88, 110, 99, 105]

plt.figure(figsize=(6, 4))
plt.plot(weeks, scores, marker='o', color='darkblue')

plt.title("Example Fantasy Team Scores over the Season")
plt.xlabel("Week")
plt.ylabel("Fantasy Points")
plt.grid(True)

plt.tight_layout()
plt.show()
```
