```python
import matplotlib.pyplot as plt

# Simulating the project's logic: How stress levels affect grades
# relative to study hours.
stress_levels = ['Low', 'Medium', 'High']
avg_grades = [88, 76, 62]

plt.figure(figsize=(8, 5))
plt.bar(stress_levels, avg_grades, color=['#4CAF50', '#FFC107', '#F44336'])

plt.title('Impact of Psychological Stress on Academic Performance', fontsize=14)
plt.xlabel('Reported Stress Level', fontsize=12)
plt.ylabel('Average Grade Percentage (%)', fontsize=12)
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

## 4. Reflection on Git History <a class="anchor" id="bullet5"></a>

**Question:** Why did the code for the Jupyter Notebook seem "foreign" in the commit history compared to the Python script?

**Answer:** The Python script (`.py`) is a **plain text file**. When you view its history, Git shows you exactly the lines of code you typed.

However, a Jupyter Notebook (`.ipynb`) is actually a **JSON (JavaScript Object Notation)** document. It contains much more than just your code; it includes:
1. **Metadata:** Information about the kernel and programming language.
2. **Cell Structure:** Tags that define whether a cell is "markdown" or "code."
3. **Outputs:** The raw data for charts and plots is often stored as long strings of characters (Base64 encoding).

Because of this hidden structure, the "raw" version of a notebook in Git looks like a complex data file rather than a simple set of instructions.

## Reflection on Commit History and Code Familiarity

When I looked at my commit history on GitHub/GitHub Desktop for the Jupyter Notebook, much of the code did **not immediately feel familiar**, even though I was the one who wrote it. This was especially true for cells that involved data cleaning, intermediate variables, or exploratory steps that were created incrementally. Because Jupyter Notebooks encourage experimentation, the final notebook reflects many small decisions made over time, which can be difficult to mentally reconstruct just by reading the code later.

In contrast, the Python script I created felt **more familiar and easier to follow**. The script was written in a more linear and intentional way, with a clear start, sequence, and end. Since Python scripts usually require planning the logic ahead of time, the structure and purpose of each section was clearer when revisiting the commit history.

The Jupyter Notebook code seemed foreign mainly because of its **nonlinear workflow**. Cells can be run out of order, variables can be redefined multiple times, and some steps may only make sense in the moment they were written. Over time, this leads to a notebook that documents *process* rather than a polished narrative. Looking back at it through commit history highlights how exploratory coding can feel disconnected from memory, even though it was essential to reaching the final result.
